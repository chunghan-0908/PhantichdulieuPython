```python
#Khai báo thư viện
import pandas as pd
import numpy as np
```

```python
#Load data
data_path = 'D:\\DA\\PhantichdulieuPython\\data\\'
data_name = 'superstore_dataset_dashboarddesign_clean.csv'
df = pd.read_csv(data_path + data_name, encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

1. Biến đổi kiểu dữ liệu

```python
#Đổi kiểu dữ liệu cho các cột Date
#Tìm các cột có tên thuộc tính chứa date
cols_contain_date = [_ for _ in df.columns.tolist() if 'Date' in _]
cols_contain_date
```

```python
#Đổi kiểu dữ liệu
for _ in cols_contain_date:
    df[_] = pd.to_datetime(df[_])
```

```python
#Kiểm tra kiểu dữ liệu
for _ in cols_contain_date:
    print('Column {} has type: {}'.format(_, df[_].dtypes))
```

2. Tạo các cột cần thiết

```python
#Tạo thêm cột Year từ trường Order Date
df['Year'] = df['Order Date'].dt.year
df.head(1)
```

```python
#Tạo cột doanh thu 
df['Revenue'] = df['Quantity'] * df['Sales']
df.head()
```

3. Gom doanh thu sản phẩm theo từng country, region và year

```python
#Sắp xếp theo region và year
df = df.sort_values(by=['Region', 'Year'])
df
```

```python
#Tính số lũy kế
df['Running Total'] = df.groupby(['Region','Year'])['Revenue'].cumsum()
df
```

```python
##Kiểm tra
#cond = df[(df['Region'] == 'Central') & (df['Year'] == 2016)]
#total_running_total = cond['Running Total'].sum()
#print(total_running_total)
```

```python
#Gom nhóm theo region, year 
running_total_by_reg_year = df.groupby(['Country','Region','Year'])['Running Total'].sum().reset_index()
running_total_by_reg_year
```

```python
# Lấy ra các cột cần thiết 
df_selected = running_total_by_reg_year.loc[:, ['Country','Region', 'Year', 'Running Total']]
df_selected
```

# Chuyển thành pivot

```python
#Chuyển thành pivot
df_selected_pivot= df_selected.pivot(index = (['Country','Region']), columns = 'Year', values = 'Running Total')
```

```python
df_selected_pivot
```

```python
#Lưu dữ liệu
data_output_path = 'D:\\DA\PhantichdulieuPython\\data_output\\'
data_output_name = 'output_processing_superstore_dataset.xlsx'

df_selected_pivot.to_excel(data_output_path + data_output_name, index = True)
```
