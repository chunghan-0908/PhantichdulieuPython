# [Bài tập] Phân tích trên bộ dữ liệu Online Retail

```python
import pandas as pd

data_path = 'D:\\DA\\PhantichdulieuPython\\data\\'
data_name = 'OnlineRetail.csv'

df = pd.read_csv(data_path + data_name, encoding = 'unicode_escape')
```

- Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
df.info()
```

- Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 

```python
avg_product_of_country = df.pivot_table(index = ['InvoiceNo', 'Country'],
                                       values = 'Quantity',
                                       aggfunc = 'mean')
avg_product_of_country
```

- Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.

```python
num_of_purchase = df.pivot_table(index = ['CustomerID','StockCode'],
                                values = 'Quantity',
                                aggfunc = ['min','max'])
```

```python
num_of_purchase
```

- Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.

```python
num_of_product = df.pivot_table(index='StockCode',
                             values=['Quantity', 'UnitPrice'],
                             aggfunc={'Quantity': 'sum', 'UnitPrice': 'mean'})
```

```python
num_of_product
```

Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
product_of_day = df.pivot_table(index = 'InvoiceDate',
                               values = 'Quantity',
                               aggfunc = 'sum')
```
