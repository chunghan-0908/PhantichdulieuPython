# Tiền xử lý

- 1. Đổi tên cột '#' thành Id cho dễ nhận dạng

```python
df = df.rename(columns={'#': 'id'})
df.head(1)
```

- 2. Đổi tên cột về dạng snake_case 

```python
new_cols = []
for _ in df.columns.tolist():
    new_cols += [re.sub('[ -]','_',_).strip().lower()]
new_cols
df.columns = new_cols  
```

```python
df.head(1)
```

- 3. Xử lý các ô bị trống giá trị 

```python
#Xử lý giá trị trống ở cột 'in_sdu_from'
df['in_sdu_from'].fillna(pd.to_datetime('2023-01-01'), inplace=True)
df.head()
```

```python
'''df['in_sdu_from'] = np.where(df['in_sdu_from'].isnull()
                      , pd.to_datetime('2023-01-01')
                      , pd.to_datetime(df['in_sdu_from'] - 25_569)*24*60*60*1e9)'''
```

```python
#Xử lý giá trị trống ở cột 'in_sdu_to'
df['in_sdu_to'].fillna(pd.to_datetime('2023-01-01'), inplace=True)
df.tail()
```

- 4. Danh sách những nhân viên nghỉ việc tạm thời

```python
list_of_absence = df[df['off_type'].notnull()].loc[:, ['id','acc', 'off_type','off_from','off_to']].reset_index()
```

```python
list_of_absence
```

```python
list_of_absence['absence_range'] = [pd.date_range(list_of_absence['off_from'][_], list_of_absence['off_to'][_]).tolist()\
                           for _ in range(list_of_absence.shape[0])]
```

```python
#Dàn dữ liệu
list_of_absence = list_of_absence.explode('absence_range')
```

```python
list_of_absence
```

```python
#Dánh dấu trạng thái là 'leave of temporary' cho những nhân viên
list_of_absence['status'] = 'Leave of temporary'
```

```python
list_of_absence.head()
```

```python
#Danh sách nhân viên làm việc trong năm 2023
df['list_of_2023'] = [pd.date_range(start = '01/01/2023', end = '12/31/2023').tolist()\
                             for _ in range(df.shape[0])]
df.head()
```

```python
#Dàn dữ liệu series 'list_of_2023' ra từng dòng
df = df.explode('list_of_2023')
```

```python
#Lấy ra các cột cần thiết
df_selected_list_of_2023 = df.loc[:, ['id','name', 'acc', 'list_of_2023']]
df_selected_list_of_2023
```

- 5. Merge danh sách ngày nghỉ với danh sách 365 ngày để tìm ra những ngày làm việc

```python
working_day = pd.merge(df_selected_list_of_2023, list_of_absence
                         , how = 'left'
                         , left_on = ['id', 'list_of_2023']
                         , right_on = ['id', 'absence_range'])
working_day.head()
```

```python
#working_day.tail(18300)
```

```python
#Xóa những bản ghi có status là 'Leave of temporary'
# .index để truy cập chỉ mục của các hàng thỏa mãn điều kiện nào đó
working_day = working_day.drop(working_day[working_day['status'] == 'Leave of temporary'].index)
```

```python
working_day.tail(18300)
```

# XỬ LÝ FILE allocation_report

```python
#Tải dữ liệu file 'allocation_report.xlsx'
all_report = pd.read_excel('allocation_report.xlsx')
```

```python
#Xem thông tin bộ dữ liệu 
all_report.info()
```

```python
#Hiểu dữ liệu
```

- 'allocation_report' là tập dữ liệu ghi lại chi tiết phân bổ làm việc của nhân viên
- Tập dữ liệu tên gồm 6 trường dữ liệu, 203 bản ghi
- Ý nghĩa các trường dữ liệu:
    + From Date: từ ngày 
    + To Date: đến ngày
    + Hours / Day: số giờ làm việc/ngày
    + Job: công việc
    + Username: tên tài khoản
    + Working For Customer: khách hàng

- 1. Đổi tên cột về dạng snake_case

```python
new_cols_report = []
for _ in all_report.columns.tolist():
    new_cols_report += [re.sub('[ -]','_',_).strip().lower()]
new_cols_report
all_report.columns = new_cols_report  
```

```python
all_report.head(1)
```

- 2. Danh sách những ngày được phân bổ công việc

```python
all_report['allocated_range'] = [pd.date_range(all_report['from_date'][_], all_report['to_date'][_]).tolist()\
                           for _ in range(all_report.shape[0])]
```

```python
all_report.head()
```

```python
#Dàn dữ liệu ra
all_report = all_report.explode('allocated_range')
```

```python
all_report.head()
```

```python
#Tạo 1 cột status đánh dấu 'Allocated'
```

```python
all_report['status'] = 'Allocated'
```

```python
all_report.head()
```

- 3. Merge 2 bảng: working_day và all_report. Để tìm ra trong khoảng thời gian làm việc nhân viên được phân bổ vào những thời gian nào

```python
allocated_by_day = pd.merge(working_day, all_report
                         , how = 'left'
                         , left_on = ['acc_x']
                         , right_on = ['username'])
allocated_by_day.head()
```

```python
# Lấy ra các cột cần thiết 
allocated_by_day_selected = allocated_by_day.loc[:, ['id','name', 'username', 'from_date', 'to_date', 'hours_/_day', 'job', 'working_for_customer','allocated_range','status_y']]
allocated_by_day_selected
```
