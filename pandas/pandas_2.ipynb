## 3. Group by and aggregation

- Chỉ sử dụng group by cho những cột có unique values nhỏ
- Luôn luôn đặt aggregration ngay sau group by

### 3.1. Examples with group by (1 column)

```python
# Mean quantity by country
df_true_price.groupby("Country")["Quantity"].mean().nlargest(10)
```

```python
# Median quantity by country
df_true_price.groupby("Country")["Quantity"].median().nsmallest(10)
```

Note: results are `Series`

```python
# Save the aggregation to result
result = df_true_price.groupby("Country")["Quantity"].median().nlargest(10)
```

result = df_true_price.groupby("Country")["Quantity"].median().nlargest(10).reset_index là để tạo thành bảng

```python
result = df_true_price.groupby("Country")["Quantity"].median().nlargest(10).reset_index()
```

```python
result
```

```python
# Verify the type
type(result)
```

```python
# Access Male index
result.loc["Netherlands"]
```

```python
# Or just like this
result["Netherlands"]
```

- More than 1 summary statistics (dùng `.agg`)

```python
# Mean and median bill by sex
df_true_price.groupby("Country")["Quantity"].agg(["min", "mean", "median", "max",  "std"])
```

- Hàm lấy phương sai

```python
(df_true_price.groupby("Country")["Quantity"].agg(["std"])**2).reset_index().rename(columns = {'std':'variance'})
```

- Hàm lấy mô tả

```python
# A shortcut
df_true_price.groupby("Country")["Quantity"].describe()
```

```python
df_true_price.groupby("Country")["Quantity"].describe()['20%']
```

```python
df_true_price["Quantity"].describe()['75%']
```

```python
df_true_price["Revenue"].describe()['75%'] - df_true_price["Revenue"].describe()['25%']
```

```python
Q3 = df_true_price["Revenue"].describe()['75%']
Q1 = df_true_price["Revenue"].describe()['25%']

Q3 - Q1
```

- Reset index để tiếp tục tính toán (nếu có)

```python
df_true_price.groupby("Country")["Quantity"].describe().reset_index()
```

### 3.2. Examples with group by (multiple columns)

```python
# Mean Quantity by Country and GoodType
df_true_price.groupby(["Country", "GoodType"])["Quantity"].mean()
```

- Note: Kết quả phía trên là một multi-index Series (2 levels)

```python
# Lưu vào result
result = df_true_price.groupby(["Country", "GoodType"])["Quantity"].mean().reset_index()
```

```python
result
```

- VD CHUYỂN THÀNH WIDE

```python
result.pivot(index = "Country", columns = "GoodType", values = "Quantity")
```

- VD CHUYỂN THÀNH LONG

Do dạng WIDE title có 2 dòng nên cần lưu vào 1 biến khác và sử dụng hàm reset_index() để chuyển về dạng data frame bình thường

```python
wide_result =  df_true_price.groupby(["Country", "GoodType"])["Quantity"].mean().reset_index()
```

```python
wide_result
```

```python
#Chuyển wide thành long
wide_result.melt(
            id_vars=["Country"],
            value_vars=["Cheap", "Luxury", "Regular"],
            var_name="GoodType",
            value_name="Quantity"
        )
```

```python
# Mean quantity for the goods that Cheap and in United Kingdom
result[("United Kingdom", "Cheap")]
```

```python
# Mean quantity for the goods that Luxury and in Singapore
result[("Singapore", "Luxury")]
```

- Ta cũng có thể reset index để được kết quả là 1 data frame

```python
result.reset_index()
```

- Tương tự như group by 1 column, ta cũng có thể dùng `.describe()`

```python
df_true_price.groupby(["Country", "GoodType"])["Quantity"].describe()
```

- Reset index

```python
df_true_price.groupby(["Country", "GoodType"])["Quantity"].describe().reset_index()
```
