```python
import pandas as pd
```

```python
#Đọc bộ dữ liệu
data_path = 'D:\\DA\PhantichdulieuPython\\data\\'
data_name = 'FoodPrice_in_Turkey.csv'

df = pd.read_csv(data_path + data_name, encoding = 'unicode_escape')
```

- Bộ dữ liệu chứa bao nhiêu dòng, bao nhiêu cột

```python
print('Bộ dữ liệu có {} dòng, {} cột'.format(df.shape[0], df.shape[1]))
```

- Thang đo tương ứng của các thuộc tính được lưu trữ

```python
df.info()
```

- Giá trung bình của từng loại thực phẩm là bao nhiêu

```python
print('Giá trung bình của từng loại thực phẩm: ')
df.groupby(['ProductId'])['Price'].mean().reset_index()
```

- Đề xuất những loại phân tích mà bạn muốn biết mà bộ dữ liệu này có thể cung cấp
