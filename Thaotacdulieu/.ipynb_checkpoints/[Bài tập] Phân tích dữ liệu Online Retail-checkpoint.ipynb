# [Bài tập] Phân tích dữ liệu Online Retail

```python
import pandas as pd

data_path = 'D:\\DA\\PhantichdulieuPython\\data\\'
data_name = 'OnlineRetail.csv'

df = pd.read_csv(data_path + data_name, encoding = 'unicode_escape')
```

- Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
df.info()
```

- Tiền xử lý dữ liệu

```python
#Đổi datatype cột InvoiceDate về dạng datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
#Lấy ra số tháng - đặt tên cột là Month
df['Month'] = df['InvoiceDate'].dt.month
df.head(10)
```

- Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;

```python
import numpy as np
```

```python
conditions = [df['Month'].isin([1,2,3]), 
              df['Month'].isin([4,5,6]), 
              df['Month'].isin([7,8,9]), 
              df['Month'].isin([10,11,12])]
choices = ['1','2','3','4']
df.loc[:, 'Previous'] = np.select(conditions, choices, default=None)
df
```

- Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice

```python
df.loc[:, 'Amount'] = df['Quantity'] * df['UnitPrice']
df
```

- Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.

```python
discount_conditions = [((df['Country'] == 'United Kingdom') & (df['Month'] == 4)),
                      (df['Country'] == 'France')]

discount_choices = ['10%', '5%']

df.loc[:, 'Discount'] = np.select(discount_conditions, discount_choices, default= '0%')
df
```

- Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.

```python
df['Discount'] = df['Discount'].astype(float)
```

```python
df['Total'] = df['Amount'] - df['Discount']
df
```
