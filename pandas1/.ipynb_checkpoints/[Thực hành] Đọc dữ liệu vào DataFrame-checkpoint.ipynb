```python
#Khai báo thư viện pandas
import pandas as pd
```

```python
df=pd.read_csv(filepath_or_buffer = 'FoodPrice_in_Turkey.csv', sep = ',')
# Hiển thị 5 dòng dữ liệu đầu tiên
df.head()
```

# [Thực hành] Chỉnh sửa cấu trúc bộ dữ liệu

- 2. Đổi tên cột thuộc tính

```python
df.rename(columns={('Month'):('Tháng'), ('Year'):('Năm')})
```

- 3. Thêm một cột mới trong DataFrame

```python
#Thêm cột mới có tên new_column và giá trị là NaN
df['new_column'] = 'NaN'
df.head()
```

```python
# Thêm cột giảm giá 10% cho tất cả các bản ghi
# Cách 1: Gán tên cột dưới dạng một chuỗi và thêm giá trị cho cột đó
df['Giảm giá']= pd.Series('5%', index=df.index)
df.head()
```

- 4. Thêm một dòng mới vào trong DataFrame

```python
#Sử dụng phương thức append()
new_rows = {'Place':'NA','ProductId':'RR','ProductName':'Rice','UmId':10,'UmName':'KG','Tháng':6,'Năm':2021,'Price':84.3785,'Giảm giá':'10%','Giảm giá 4':'10%','Giảm giá 2':'12%', 'new_column':'NaN'}
df = df.append(new_rows, ignore_index=True)
df.tail()
```

- 5. Xóa một cột trong DataFrame

```python
#Xóa 1 cột
del df['new_column']
```

```python
#Sử dụng phương thức pop()
df.pop('Giảm giá')
```

```python
#Sử dụng phương thức drop()
df.drop(columns=['Giảm giá 4'], inplace=True)
```

```python
df.head()
```

- 6. Xóa các dòng trong DataFrame

```python
df.drop(10, axis = 0, inplace=True)
df
```

```python
# Xóa các dòng có chỉ số 7277 và 7279 sử dụng phương thức drop()
df.drop([7277,7279], inplace=True)
df
```
