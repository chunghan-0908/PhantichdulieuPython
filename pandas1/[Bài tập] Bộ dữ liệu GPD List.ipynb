```python
import pandas as pd

data_path = 'D:\\DA\\PhantichdulieuPython\\data\\'
data_name = 'GDPlist.csv'

df = pd.read_csv(data_path + data_name, encoding = 'unicode_escape')
```

- ƒê·ªçc b·ªô d·ªØ li·ªáu, cho bi·∫øt s·ªë d√≤ng, s·ªë c·ªôt v√† ki·ªÉu d·ªØ li·ªáu c·ªßa c√°c thu·ªôc t√≠nh.

```python
df.info()
```

- Vi·ªát h√≥a t√™n c√°c c·ªôt trong b·∫£ng d·ªØ li·ªáu: Country ü°™ Nuoc; Continent ü°™ Chauluc; GDP (millions of US$) ü°™ GDP (trieu $)

```python
data = df.rename(columns={('Country'):('Nuoc'), ('Continent'):('Chauluc'), ('GDP (millions of US$)'):('GDP(trieu)')}).copy()
```

- Ch√®n th√™m m·ªôt c·ªôt ‚ÄúThanhpho‚Äù v√†o sau c·ªôt ‚ÄúNuoc‚Äù, gi√° tr·ªã ban ƒë·∫ßu l√† gi√° tr·ªã c·ªßa c·ªôt ‚ÄúNuoc‚Äù    

```python
#T√¨m v·ªã tr√≠ c·ªßa c·ªôt Nuoc
col_index =  data.columns.get_loc('Nuoc')
```

```python
data.insert(col_index + 1, 'Thanhpho', data.loc[:, 'Nuoc'])
```

- Trong c·ªôt Thanhpho, thay gi√° tr·ªã Vietnam th√†nh Hanoi; L√†m t∆∞∆°ng t·ª± v·ªõi c√°c n∆∞·ªõc c√≤n l·∫°i.
