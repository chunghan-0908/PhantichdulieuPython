# [Thực hành] Truy cập dữ liệu trong DataFrame

```python
#Khai báo thư viện pandas
import pandas as pd
```

```python
df=pd.read_csv(filepath_or_buffer = 'FoodPrice_in_Turkey.csv', sep = ',')
# Hiển thị 5 dòng dữ liệu đầu tiên
df.head()
```

```python
df.info()
```

- 2. Sử dụng phương thức iloc để truy cập dữ liệu

```python
# Truy cập các dòng có chỉ số liên tục từ 3 tới 8 của dữ liệu
df.iloc[3:8]
df
```

```python
# Truy cập dòng có chỉ số 2 của dữ liệu
data = df.iloc[2]
data
```

```python
# Truy cập các dòng có chỉ số liên tục từ 3 tới 8 của dữ liệu
data = df.iloc[3:8]
data
```

```python
# Truy cập các dòng rời rạc của dữ liệu
data = df.iloc[[3,5,7]]
data
```

```python
# Truy cập cột thứ 4 của dữ liệu
# : được sử dụng để đại diện cho tất cả các dòng.
data = df.iloc[:,4]
data
```

```python
# Truy cập các cột liên tục từ cột 3 tới cột 9 của dữ liệu
data = df.iloc[:,3:9]
data
```

```python
# Truy cập tới phần tử tại dòng 3 cột 7 của dữ liệu
data = df.iloc[3,7]
data
```

```python
# Truy cập tới các phần tử từ dòng 3 đến dòng 4, cột 5 đến cột 6 của dữ liệu
data = df.iloc[3:5,5:7]
data
```

- 3. Sử dụng phương thức loc để truy cập dữ liệu
loc có thể làm việc được với chỉ số hàng, tên cột và các biểu thức điều kiện của cột

```python
# Truy cập dòng có chỉ số 3 của dữ liệu
data = df.loc[3]
data
```

```python
# Truy cập cột thứ 4 của dữ liệu
data = df.loc[:,'UmName']
data
```

```python
# Truy cập cột thứ 4,5 của dữ liệu
data = df.loc[:,['UmName','Month']]
data
```

```python
# Truy cập tới phần tử tại dòng 3 cột 7 của dữ liệu
data = df.loc[3,'Price']
data
```

```python
# Truy cập tới phần tử có Year >=2019
data = df.loc[df.Year >= 2019].copy()
data
```

- 4. Thay thế các giá trị trong DataFrame

```python
# Thay số 5 bằng số 10 trong toàn bộ dữ liệu
df.replace(5,10,inplace = True)
df.head()
```

```python
# Thay mã sản phẩm từ giá trị 52 thành RR trong toàn bộ dữ liệu
# RR là chuỗi ký tự nên được để trong cặp dấu nháy đơn
df.replace(52,'RR',inplace = True)
df.head()
```

```python
# Thay giá trị 10 trong cột Month thành giá trị 5
df['Month'].replace(10,5,inplace = True)
df.head()
```
